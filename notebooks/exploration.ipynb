```python
import uproot
import pandas as pd
from uproot_methods.classes.TLorentzVector import TLorentzVector
import tqdm
import numpy as np
```

```python
f = uproot.open('../output_H_to_WlnuWlnu.root')
```

```python
f.keys()
```

```python
f['event;1'].keys()
```

```python
events = f['event;1']
```

```python
len(events.keys())
```

```python
df = pd.DataFrame()
```

```python
for k, v in events.items():
    name = k.decode('ascii')
    values = v.array()
    if type(values[0]) is TLorentzVector:
        x = np.zeros(values.shape)
        y = np.zeros(values.shape)
        z = np.zeros(values.shape)
        E = np.zeros(values.shape)
        m = np.zeros(values.shape)
        for i, vec in tqdm.tqdm(enumerate(values), total=len(values)):
            x[i] = vec.x
            y[i] = vec.y
            z[i] = vec.z
            E[i] = vec.t
            try:
                m[i] = vec.mass
            except:
                m[i] = 0.0
        df[f'{name}x'] = x
        df[f'{name}y'] = y
        df[f'{name}z'] = z
        df[f'{name}E'] = E
        df[f'{name}m'] = m
    else:
        df[name] = values
```
